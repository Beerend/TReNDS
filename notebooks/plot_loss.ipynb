```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
model  = 'resnet10/0'
filename  = '../results/'+model+'/log.txt'
losses    = pd.read_csv('test_log.txt')
max_epoch = np.max(losses['Epoch'].to_list())
losses    = [losses.loc[losses['set'] == 'train'], losses.loc[losses['set'] == 'test']]

fig, axes = plt.subplots(2, 1, figsize=(10,5))
for i, row in enumerate(axes):
    ax = sns.lineplot(x='Epoch', y='MAE', data=losses[i], ax=row)
    ax.set_ylabel('Loss (MAE)', weight='bold')
    ax.set_ylim((0,50))
    ax.set_xlabel('Epoch', weight='bold')
    ax.set_xticks(list(range(1,max_epoch+1)))
    
fig.tight_layout()
plt.show()
```
